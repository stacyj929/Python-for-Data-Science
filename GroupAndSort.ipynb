```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reviews = pd.read_csv("winemag-data_first150k.csv", index_col=0)
```

```python
reviews.sample(5)
```

```python
reviews.info()
```

```python
# sort by index
reviews.sort_index(ascending=False)
```

```python
# sort by column province in decending order
reviews.sort_values(by='province', ascending=False).head(5)
```

```python
# groupby()
reviews.groupby('country').province.count()
```

```python
reviews.groupby(['country','province']).points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
# locate wine with 90 points and price of 5
reviews.loc[(reviews['points'] == 90) & 
            (reviews['price'] == 5)]
```

```python
# grouped by winery and used lambda function to pick out the first row
reviews.groupby('winery').apply(lambda df: df.iloc[0:1])
```

```python
reviews.winery.value_counts()
```

```python
# group by winery and show the wine with the highest points
reviews.groupby('winery').apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# group by country and show the number of records, max and min prices
reviews.groupby('country').price.agg([len,max,min])
```

```python
# number of records grouped by province and country
reviews.groupby(['country','province']).size()
```

```python
# check if reviews has an index and what type
type(reviews.index)
```

```python
# create a multi-indexed df
reviews_CP = reviews.groupby(['country','province']).points.mean()
```
