```python
col_names = ['Bahamas','Jamaica','California','Denmark']
my_df = pd.DataFrame(data= [[1,2,3,4],[4,5,6,6],[5,8,9,4]],
                     index=range(0,3), columns=col_names)
```

```python
my_df
```

```python
# return row 1 from df
my_df.iloc[[1]]
```

```python
# get column jamaica - position ref 
my_df.iloc[:,[1]]
```

```python
# get columns 1 and 2 - positional ref
# my_df.iloc[:,[0,1]]
my_df.iloc[:,0:2]
```

```python
# get column cali - names
my_df.loc[:,['Bahamas','California']]
```

```python
# get top two rows for all columns 

my_df.iloc[0:2,:]
```

```python
# get top two rows for all columns - names
my_df.loc[0:1,:]
```

```python
# last two rows of Cali and Denmark - positional ref
my_df.iloc[-2:,[2,3]]
```

```python
# last two rows of Cali and Denmark - names 
my_df.loc[[1,2],['California','Denmark']]
```

```python
# change row data - positional ref
# my_df.iloc[[2]] = [50,80,90,40]
my_df.iloc[2]
```
