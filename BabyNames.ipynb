```python
# load data from all files
names_all=[]
for year in range(1880, 2018):
    names_all.append(pd.read_csv('data/names/yob{}.txt'.format(year), 
                                 names=['name','sex','number']))
    names_all[-1]['year'] = year 
allyears = pd.concat(names_all)
allyears.head()
```
