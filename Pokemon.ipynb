Pokemon plotting

```python
# download the pokemon.csv file from blackboard
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pokemon = pd.read_csv("data/pokemon.csv")
```

```python
pokemon.info()
```

```python
pokemon['type1'].value_counts().plot(kind='bar')
```

```python
# plot hp (hit points)
pokemon['hp'].describe()
```

```python
# line plot of hp
# pokemon['hp'].value_counts().plot(kind='line')
pokemon['hp'].value_counts().sort_index().plot(kind='line')
```

```python
pokemon['weight_kg'].plot(kind='hist')
```

```python
pokemon.plot(kind='scatter', x='attack', y='defense')
```

```python
pokemon.plot(kind='hexbin', x='attack', y='defense', gridsize=15)
```

```python
# plot attack rating and defense rating to compare
fig, axarr = plt.subplots(2,1, figsize=(8,8))
pokemon['attack'].plot(kind='hist', ax=axarr[0], title='pokemon attack rating')
pokemon['defense'].plot(kind='hist', ax=axarr[1], title='pokemon defense rating')
```
